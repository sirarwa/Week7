```python
# Using the Pandas library for data manipulation and analysis
import pandas as pd
# Read the CSV file into a Pandas DataFrame      
df = pd.read_csv('Walmart_Sales.csv')

# Specify the categorical column for grouping
categorical_column = 'Holiday_Flag' 

# Specify the numerical column for which to compute the mean
numerical_column = 'Temperature' 

# Perform the grouping and compute the mean
grouped_means = df.groupby(categorical_column)[numerical_column].mean()

# Display the resulting grouped means
print(f"Mean {numerical_column} by {categorical_column}:")
print(grouped_means)
```

```python
# Loading matplotlib library for visualization.
import matplotlib.pyplot as plt

# Read the Sales data .csv file
df = pd.read_csv('Walmart_Sales.csv')

# Convert 'Date' column to datetime and specify the format.
df['Date'] = pd.to_datetime(df['Date'], format='mixed')

# Group sales by month ready for plotting
monthly_sales = df.groupby(pd.Grouper(key='Date', freq='ME'))['Weekly_Sales'].sum().reset_index()

# Create the line chart for sales between 2010 to 2013 as per the csv file.
plt.figure(figsize=(10, 6)) # Set the figure size
plt.plot(monthly_sales['Date'], monthly_sales['Weekly_Sales'], color='orange') # Set the color of the line to orange
plt.title('MONTHLY SALES FROM 2010 TO 2013', color='green') # Set the title of the plot
plt.xlabel('Month', color='green') # Set the x-axis label
plt.ylabel('Sales (Millions)') # Set the y-axis label
plt.grid(True) # Add grid lines to the plot
plt.show() # Display the plot
```

```python
#Plotting a bar chart where each bar represents a store, and the height of the bar represents the average weekly sales for that store.
# using the pandas library for data manipulation and analysis.
import pandas as pd
# Using the matplotlib library for plotting
import matplotlib.pyplot as plt

# Loading the s
df = pd.read_csv('Walmart_Sales.csv')

# Grouping the data by 'Store' and calculating the mean of 'Weekly_Sales'
store_sales = df.groupby('Store')['Weekly_Sales'].mean().reset_index()

# Create the bar chart
plt.figure(figsize=(15, 6))  # Adjust figure size as needed
plt.bar(store_sales['Store'], store_sales['Weekly_Sales'], color='indigo') #Plot and set the bar properties
plt.title('Average Weekly Sales by Store', color='blue') #The title of the bar chart
plt.xlabel('Store', color='blue') #Plot store number on the y-axis
plt.ylabel('Average Weekly Sales (Millions)', color='blue') #Plot avg weekly sales on the y-axis
plt.xticks(store_sales['Store']) #Ensure all store names are displayed
plt.show() #Display the bar chart
```

```python
# Using the Pandas library for data manipulation and analysis and matplotlib for plotting.
import matplotlib.pyplot as plt

# Let's say you want to see the distribution of 'Weekly_Sales'
Sales = 'Weekly_Sales'

# Creating the histogram
# Set the figure size
plt.figure(figsize=(10, 6)) 
# Set the number of bins and color for the histogram
plt.hist(df[Sales], bins=20, edgecolor='black', color='magenta') # Set the title and labels for the histogram
plt.title('Distribution of Weekly Sales', color='blue') # Set the x and y labels
plt.xlabel('Weekly Sales', color='blue') # Set the y-axis label
plt.ylabel('Frequency', color='blue') # Add grid lines for better readability
plt.grid(axis='y', alpha=0.75) # Add a vertical line for the mean
plt.show() # Display the histogram
```
